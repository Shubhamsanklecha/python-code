```python
person1 = input("choose any one from stone, paper and scissor: ")
person2 = input("choose any one from stone, paper and scissor: ")
if person1 == "stone" and person2 == "scissor" or person1 == "paper" and person2 == "stone" or person1 == "scissor" and person2 == "paper":
    print("Person1 is the Winner!!!")
elif person1 == person2:
    print("its a Tie!!")
elif person1 == "scissor" and person2 == "stone" or person1 == "stone" and person2 == "paper" or person1 == "paper" and person2 == "scissor":
    print("Person2 is the Winner!!!")
else:
    print("Sorry!! Wrong input")
```
