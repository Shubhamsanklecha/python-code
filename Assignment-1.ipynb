### Write a python program to convert only odd indexed characters to lower case 
### Ex:- PYTHON
### Output: - PyThOn

```python
variable = input("enter anything ")
dummy=""
for x in range(len(variable)):
    if x%2==0:
        dummy = dummy+(variable[x])
    else:
        dummy= dummy+(variable[x]).lower()
print(dummy)
    
```
