### Write a Python program which iterates the integers from 1 to 60. For multiples of three print "Fizz" instead of the number and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".

```python
# using for loop
b=[]
for a in range(1,61):
    if a%3==0 and a%5==0:
        b.append("fizzBuzz")
    elif a%3==0:
        b.append("Fizz")
    elif a%5==0:
        b.append("Buzz")
    else:
        b.append(a)
print(b)
```

```python
# using while loop
x=True
b=[]
while x:
    for a in range(1,61):
        if a%3==0 and a%5==0:
            b.append("fizzBuzz")
        elif a%3==0:
            b.append("Fizz")
        elif a%5==0:
            b.append("Buzz")
        else:
            b.append(a)
        x=False
    print(b)
```

### Write a Python program that accepts a string and calculate the number of digits and letters
Example:
Input
Python 3.2 
Output
Letters 6
Digits 2

```python
# using for loop
str1=input("enter anything ")
digits=0
letters=0
for a in str1:
    if a.isdigit():
        digits = digits+1
    elif a.isalpha():
        letters=letters+1
print("Digits= ",digits)
print("Letters= ",letters)
```

```python
# using while loop
str1=input("enter anything ")
digits=0
letters=0
x=True
while x:
    for a in str1:
        if a.isdigit():
            digits = digits+1
        elif a.isalpha():
            letters=letters+1
        x=False
    print("Digits= ",digits)
    print("Letters= ",letters)
```

# Write a Python program to check the validity of password input by users. 
##### Validation:
##### At least 1 letter between [a-z] and 1 letter between [A-Z].
##### At least 1 number between [0-9]. 
##### At least 1 character from [# @].
##### Minimum length 6 characters.
##### Maximum length 16 characters.

```python
# using for loop
password=input("enter your password: ")
Alpha=0
alpha=0
digits=0
symbol=0
for a in password:
    if a.isalpha():
        if a==a.upper():
            Alpha=Alpha+1
        elif a==a.lower():
            alpha=alpha+1
    elif a.isdigit():
        digits=digits+1
    elif a=="@" or a=="#" or a=="$":
        symbol=symbol+1
if len(password)<6:
    print("minimum 6 characters required!!!")
elif len(password)>16:
    print("maximum 16 characters are allowed!!!")
elif Alpha==0:
    print("Atleast 1 uppercase letter is needed")
elif alpha==0:
    print("Atleast 1 lowercase letter is needed")
elif digits==0:
    print("Atleast 1 digit is required")
elif symbol==0:
    print("Atleast 1 character is required from @,#,$")
else:
    print("You have entered correct password")
```

### Write a program which takes integer inputs from user until he/she presses “q” (Ask to press q to quit after every integer input). Print average and product of all numbers.

```python
# using while loop
x="a"
prod=1
total=0
count=0
while x!="q":      
    num=int(input("Type the integer: "))
    prod=prod*num
    total=total+num
    count=count+1
    x=input("Type q to quit: ")
print("Average of all the numbers is: ",total/count)
print("Product of all the numbers is: ",prod)
```

### Create a dynamic calculator which will run continually till you press “c”

```python
# using while loop
while num1 or num2 != "c" or "C": 
    num1 = int(input("enter first number "))
    num2 = int(input("enter second number"))
    var = input("enter the symbol what you want to perform ")
    if var =="+":
        print(num1+num2)
    elif var =="-":
        print(num1-num2)
    elif var =="*":
        print(num1*num2)
    elif var =="/":
        print(num1/num2)
    elif var =="%":
        print(num1%num2)
    elif var =="//":
        print(num1//num2)
    elif var =="**":
        print(num1**num2)
    elif var=="c" or "C":
        break
    else:
        print("you have entered wrong character")
```
